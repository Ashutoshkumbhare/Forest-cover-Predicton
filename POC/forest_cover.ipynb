The dataset looks imbalance. Let's use SMOTE to balance the dataset.

But first let's concat our numerical dataframe and categorical dataframe.

let's say we have dataset in which we have some classes, data in the dataset should not be less or more for each class,
   means if we have more data on class 1,2 and less on class 3,4 in the dataset model may not give best accuracy for the
   class having less data to learn.(use SMOTE).

```python
# let's concat two dataframes for categorical and numerical data together

final_data = pd.concat([num_data1,cat_data],axis=1)
final_data
```

```python
X = final_data.drop(['class'],axis=1)
y = data['class']
```

### What is SMOTE? 
Just like the name suggests, the technique generates synthetic data for the minority class. SMOTE proceeds by joining the points of the minority class with line segments and then places artificial points on these lines.
when ever the values are not present it trys to copy previous values and it will trys to increase the rows.

```python
#!pip install imblearn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sample = SMOTE()
```

```python
X_bal,y_bal = sample.fit_resample(X,y)
```

```python
plot.figure(figsize=(10,5), facecolor='white')
sns.countplot(y_bal)
```

### Let's create Clusters.

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=69)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
print(wcss)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of cluster')
plt.ylabel('WCSS')
plt.show()
```

Look's like elbow is between 2-4.Let's take 3 as cluster_number

```python
cluster = KMeans(n_clusters=3,init="k-means++",n_init=5,random_state=8)
cluster_predict = cluster.fit_predict(X)
```

```python
print(cluster_predict)
```

Let's appent cluster array as column in the X dataset

```python
X['cluster'] = cluster_predict
X['Labels'] = y
```

### 2. Randomforest algo.
Article: https://www.analyticsvidhya.com/blog/2015/06/tuning-random-forest-model/

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_params = {"n_estimators": [10, 50, 100, 130], "criterion": ['gini', 'entropy'],
                               "max_depth": range(2, 4, 1), "max_features": ['auto', 'log2']}
```

```python
RF_obj=RandomForestClassifier()
RF_grid = GridSearchCV(estimator=RF_obj, param_grid=randomforest_params, cv=5,  verbose=3,n_jobs=-1)
```

```python
RF_grid.fit(X_train,Y_train)
```

```python
RFcriterion = RF_grid.best_params_['criterion']
RFmax_depth = RF_grid.best_params_['max_depth']
RFmax_features = RF_grid.best_params_['max_features']
RFn_estimators = RF_grid.best_params_['n_estimators']
```

```python
RF_new =RandomForestClassifier(n_estimators=RFn_estimators,criterion=RFcriterion,max_depth=RFmax_depth,max_features=RFmax_features)
```

```python
RF_new.fit(X_train,Y_train)
```

```python
RF_pridect_Y = RF_new.predict(X_test)
```

```python
RFaccuracy_score = accuracy_score(Y_test,RF_pridect_Y)
RFaccuracy_score
```

#### Randomforest score is also not so good.
